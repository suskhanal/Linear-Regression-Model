```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('income-rent.csv')
data
```

```python
X = data[['Income']]
y = data[['Rent']]
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(10, 4))
sns.regplot(x = 'Income', y = 'Rent', data = data)
plt.xticks(range(1200, 5500, 100), rotation = 90)
plt.grid()
plt.show()
```

```python
# Create a Linear Regression Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)     # train the model according to the data
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict( [[3300]] )
```

```python
# Testing the model
y_pred = model.predict(X)
y_pred
```

```python
data['Pred'] = np.round(y_pred, 1)
data
```

```python
from sklearn.metrics import r2_score
np.round(r2_score(data['Rent'], data['Pred']), 4)
```

```python
np.round(r2_score(y, y_pred), 4)
```
